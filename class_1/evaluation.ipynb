```python
#importing the libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
```

```python
df = pd.read_csv('output_file1.csv')
```

```python
# Check for unique values in 's', 'time_taken_ms', and 'n_compares'
print(df['s'].value_counts())
print(df[['n_compares', 'time_taken_ms']].describe())

# Plot without outliers to see if that changes the view
plt.figure(figsize=(12, 6))
sns.boxplot(x='s', y='time_taken_ms', data=df, showfliers=False)
plt.title('Box Plot of Time Taken (ms) for Each Value of s (Without Outliers)')
plt.xlabel('s')
plt.ylabel('Time Taken (ms)')
plt.grid(True)
plt.show()

# Similarly, try plotting n_compares without outliers
plt.figure(figsize=(12, 6))
sns.boxplot(x='s', y='n_compares', data=df, showfliers=False)
plt.title('Box Plot of Number of Comparisons for Each Value of s (Without Outliers)')
plt.xlabel('s')
plt.ylabel('Number of Comparisons')
plt.grid(True)
plt.show()
```

```python
# Grouping the data by 's' and calculating the mean time for each group
mean_time_by_s = df.groupby('s')['time_taken_ms'].mean()
print(mean_time_by_s)

# Extracting the mean times for each value of 's'
mean_time_s98 = mean_time_by_s[98]
mean_time_s3 = mean_time_by_s[3]
mean_time_s1 = mean_time_by_s[1]

# Calculating the absolute percentage difference to ensure it's not negative
perc_diff_s98_vs_s3 = abs((mean_time_s98 - mean_time_s3) / mean_time_s3) * 100
perc_diff_s98_vs_s1 = abs((mean_time_s98 - mean_time_s1) / mean_time_s1) * 100

print(f"Percentage difference between s=98 and s=3: {perc_diff_s98_vs_s3:.2f}%")
print(f"Percentage difference between s=98 and s=1: {perc_diff_s98_vs_s1:.2f}%")
```

```python
# Grouping the data by 's' and calculating the mean time and number of comparisons for each group
mean_time_by_s = df.groupby('s')['time_taken_ms'].mean()
mean_compares_by_s = df.groupby('s')['n_compares'].mean()

# Extracting the mean times and comparisons for each value of 's'
mean_time_s3 = mean_time_by_s[3]
mean_time_s1 = mean_time_by_s[1]

mean_compares_s98 = mean_compares_by_s[98]
mean_compares_s3 = mean_compares_by_s[3]
mean_compares_s1 = mean_compares_by_s[1]
```

```python
# Calculating the absolute percentage difference for number of comparisons between s = 1 and s = 3
perc_diff_compares_s3_vs_s1 = abs((mean_compares_s3 - mean_compares_s1) / mean_compares_s1) * 100

# Calculating the absolute percentage difference for time taken between s = 1 and s = 3
perc_diff_time_s3_vs_s1 = abs((mean_time_s3 - mean_time_s1) / mean_time_s1) * 100

# Calculating the absolute percentage difference for number of comparisons between s = 1 and s = 98
perc_diff_compares_s98_vs_s1 = abs((mean_compares_s98 - mean_compares_s1) / mean_compares_s1) * 100

# Printing the results
print(f"Percentage difference in number of comparisons between s=1 and s=3: {perc_diff_compares_s3_vs_s1:.2f}%")
print(f"Percentage difference in time taken between s=1 and s=3: {perc_diff_time_s3_vs_s1:.2f}%")
print(f"Percentage difference in number of comparisons between s=1 and s=98: {perc_diff_compares_s98_vs_s1:.2f}%")
```

```python
# Grouping the data by 's' and calculating the mean time and number of comparisons for each group
mean_time_by_s = df.groupby('s')['time_taken_ms'].mean()
mean_compares_by_s = df.groupby('s')['n_compares'].mean()

# Extracting the mean times and comparisons for each value of 's'
mean_time_s3 = mean_time_by_s[3]
mean_time_s1 = mean_time_by_s[1]

mean_compares_s98 = mean_compares_by_s[98]
mean_compares_s3 = mean_compares_by_s[3]
mean_compares_s1 = mean_compares_by_s[1]

# Calculating the absolute percentage difference for number of comparisons
perc_diff_compares_s3_vs_s1 = abs((mean_compares_s3 - mean_compares_s1) / mean_compares_s1) * 100
perc_diff_compares_s98_vs_s1 = abs((mean_compares_s98 - mean_compares_s1) / mean_compares_s1) * 100

# Calculating the absolute percentage difference for time taken
perc_diff_time_s3_vs_s1 = abs((mean_time_s3 - mean_time_s1) / mean_time_s1) * 100

# Printing the results
print(f"Mean number of comparisons for s=1: {mean_compares_s1:.2f}")
print(f"Mean number of comparisons for s=3: {mean_compares_s3:.2f}")
print(f"Mean number of comparisons for s=98: {mean_compares_s98:.2f}")

print(f"Percentage difference in number of comparisons between s=1 and s=3: {perc_diff_compares_s3_vs_s1:.2f}%")
print(f"Percentage difference in number of comparisons between s=1 and s=98: {perc_diff_compares_s98_vs_s1:.2f}%")

print(f"Mean time taken for s=1: {mean_time_s1:.2f} ms")
print(f"Mean time taken for s=3: {mean_time_s3:.2f} ms")

print(f"Percentage difference in time taken between s=1 and s=3: {perc_diff_time_s3_vs_s1:.2f}%")
```
